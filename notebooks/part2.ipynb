```python
import pandas as pd
import matplotlib.pyplot as plt

# import datasets
bcn = pd.read_csv("../data/bcn_listings.csv.gz")
hk = pd.read_csv("../data/Hong_Konglistings.csv.gz")
```

```python
# Select the "bedrooms" column
bcn_bedrooms = bcn["bedrooms"].dropna()
hk_bedrooms = hk["bedrooms"].dropna()
```

```python
# Convert to integers
bcn_bedrooms = bcn_bedrooms.astype(int)
hk_bedrooms = hk_bedrooms.astype(int)
```

```python
# Create the plot
plt.figure(figsize=(10,6))
plt.hist(
    bcn_bedrooms, 
    bins=range(0,8), 
    alpha=0.5, 
    label="Barcelona",
    color="#AEC6CF"
)
plt.hist(
    hk_bedrooms, 
    bins=range(0,8), 
    alpha=0.5, 
    label="Hong Kong",
    color="#F7C6D4"
)

plt.xlabel("Number of Bedrooms")
plt.ylabel("Number of Listings")
plt.title("Distribution of Number of Bedrooms: Barcelona vs Hong Kong")
plt.legend()
```

```python
# Save the graph
plt.savefig("../reports/graph3.png", dpi=300)

plt.show()
```

```python
# another example about room type distribution 
bcn_rt = bcn["room_type"].value_counts(normalize=True)
hk_rt  = hk["room_type"].value_counts(normalize=True)
```

```python
df = pd.DataFrame({
    "Barcelona": bcn_rt,
    "Hong Kong": hk_rt
})
df.plot(
    kind="bar", 
    figsize=(10,6),
    color=["#6DB4CD", "#F28EAB"]
)
plt.title("Room Type Distribution")
plt.ylabel("Proportion")
plt.xlabel("Room Type")
```
